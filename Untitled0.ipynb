# Entrenar y evaluar el modelo


```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Cargar el conjunto de datos Titanic
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')

# Preprocesamiento de datos
# Llenar valores nulos en columnas numéricas relevantes
numeric_columns = ['Age', 'Fare']  # Ajusta esto según las columnas numéricas en tus datos
train[numeric_columns] = train[numeric_columns].fillna(train[numeric_columns].mean())
test[numeric_columns] = test[numeric_columns].fillna(test[numeric_columns].mean())

# Codificar variables categóricas en variables dummy
train = pd.get_dummies(train, columns=['Sex', 'Embarked'], drop_first=True)
test = pd.get_dummies(test, columns=['Sex', 'Embarked'], drop_first=True)

# Definir las características (features) y etiquetas (labels)
X_train = train.drop(['Survived', 'Name', 'Ticket', 'Cabin'], axis=1)
y_train = train['Survived']
X_test = test.drop(['Name', 'Ticket', 'Cabin'], axis=1)

# Dividir el conjunto de datos en entrenamiento y prueba
# Nota: No se divide X_test ya que se utilizará para predecir en el conjunto de datos de prueba
X_train_split, X_val_split, y_train_split, y_val_split = train_test_split(X_train, y_train, test_size=0.2, random_state=42)

# Verificar las formas de los datos
print(f'X_train_split shape: {X_train_split.shape}')
print(f'X_val_split shape: {X_val_split.shape}')
print(f'y_train_split shape: {y_train_split.shape}')
print(f'y_val_split shape: {y_val_split.shape}')
print(f'X_test shape: {X_test.shape}')
```
