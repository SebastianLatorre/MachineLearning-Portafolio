## Pregunta 7: Predicción con modelos de ML


Respuesta: Primero comparamos con DummyClassifier y luego utilizaremos Knn Vecinos cercanos y Random Forest

#### DummyClassifier

```python
from sklearn.dummy import DummyClassifier
from sklearn.metrics import accuracy_score

# Crear el DummyClassifier con la estrategia 'most_frequent'
dummy_clf = DummyClassifier(strategy='most_frequent')

# Entrenar el DummyClassifier
dummy_clf.fit(X_train, y_train)

# Hacer predicciones en el conjunto de prueba
y_pred = dummy_clf.predict(X_test)

# Evaluar el rendimiento del modelo
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')
```

#### KNeighborsClassifier

```python
from sklearn.neighbors import KNeighborsClassifier
import sklearn.metrics as metrics
```

Este codigo fue sacado de los Colab vistos en Clases. Aqui buscamos encontrar el mejor accuracy sin sobreentrenar

```python
K = 10
max_accuracy = 0
best_neighbors = 0
for n_neighbors in range(1, K + 1):
    print(f"Evaluando rendimiento para {n_neighbors} vecinos")
    # Definimos y ajustamos nuestro clasificador sobre los datos de entrenamiento
    classifier = KNeighborsClassifier(n_neighbors=n_neighbors).fit(
        X_train, y_train
    )
    # Usamos el clasificador para predecir sobre los datos de test
    predicted = classifier.predict(X_test)
    accuracy = metrics.balanced_accuracy_score(y_test, predicted)
    print(f"K={n_neighbors}: {accuracy:.2f}")
    if accuracy > max_accuracy:
        max_accuracy = accuracy
        best_neighbors = n_neighbors
print(
    f"El mejor rendimiento se alcanza con {best_neighbors} vecinos "
    f"junto con un accuracy de {max_accuracy}."
)
```

Evaluando el modelo escogemos 4 vecinos cercanos
`K=4`: con un `accuracy de 0.76` ya que despues de esto podemos sobre-entrenar el modelo. Tambien obtenemos la matriz de confusión para este mismo modelo donde se equivoco `37` veces en `"0"` y `91` veces en `"1"` dando como resultado el accuracy mencionado

```python
# Definimos y ajustamos nuestro clasificador sobre los datos de entrenamiento
classifier = KNeighborsClassifier(n_neighbors=4)
classifier.fit(X_train, y_train)

# Usamos el clasificador para predecir sobre los datos de test
predicted = classifier.predict(X_test)

expected = y_test
print("Confusion matrix:\n%s" % metrics.confusion_matrix(expected, predicted))
print()
print("Accuracy: %s" % metrics.balanced_accuracy_score(expected, predicted))
```

#### RandomForestClassifier

En estre código entrenamos un modelo de `random forest` donde tenemos una cantidad de `100 arboles` aleatorios evaluamos el `accuracy` de este modelo obteniendo `85,22%` de rendimiento. Ademas utilizamos la matriz de confusion para saber donde fue que mas se equivoco y donde fue que acertó mas. en este caso se equivoco `31` veces en `"0"` y `45` veces en `"1"`

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
from sklearn.metrics import balanced_accuracy_score, classification_report, confusion_matrix

forest = RandomForestClassifier(n_estimators=100, random_state=0) # Cantidad de árboles en el "forest"
forest.fit(X_train, y_train)
predicted = forest.predict(X_test)

print("Matriz de confusión:")
print(confusion_matrix(y_test, predicted))
print()

print("Accuracy:")
print(balanced_accuracy_score(y_test, predicted))
print()

print("Classification report:")
print(classification_report(y_test, predicted))
```

## Pregunta 8: Sobreentrenamiento


Respuesta:

Como se puede observar en el gráfico generado a continuación al igual que en la pregunta anterior es en K=4 donde alcanzamos el mejor Accuracy y a la vez nos podemos dar cuenta de que si seguimos avanzando con k=5,6,7,8... estaremos sobre-entrenando nuestro modelo e incluso alcanzando peor desempeño (Accuracy) y perdiendo capacidad de generalización en algunos casos (k=5,6,7,8,9)

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import balanced_accuracy_score

# Listas para almacenar los resultados
train_scores = []
test_scores = []

# Rango de números de vecinos
n_neighbors_range = range(1, 21)

# Entrenar y evaluar el modelo para cada número de vecinos
for n_neighbors in n_neighbors_range:
    clf = KNeighborsClassifier(n_neighbors=n_neighbors)
    clf.fit(X_train, y_train)

    # Evaluar en el conjunto de entrenamiento
    y_train_pred = clf.predict(X_train)
    train_score = balanced_accuracy_score(y_train, y_train_pred)
    train_scores.append(train_score)

    # Evaluar en el conjunto de prueba
    y_test_pred = clf.predict(X_test)
    test_score = balanced_accuracy_score(y_test, y_test_pred)
    test_scores.append(test_score)

# Graficar los resultados
plt.figure(figsize=(12, 5))

plt.plot(n_neighbors_range, train_scores, label='Train', marker='o')
plt.plot(n_neighbors_range, test_scores, label='Test', marker='o')
plt.xlabel('Número de vecinos')
plt.ylabel('Balanced Accuracy')
plt.title('Sobreentrenamiento a medida que aumenta el número de vecinos en KNN')
plt.legend()
plt.grid(True)
plt.show()
```
