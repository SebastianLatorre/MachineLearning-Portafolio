````python
import networkx as nx
import matplotlib.pyplot as plt
from networkx.algorithms.isomorphism import faster_could_be_isomorphic, is_isomorphic

# Crear los grafos A, B y C
G_A = nx.Graph()
G_B = nx.Graph()
G_C = nx.Graph()

# Grafo A: un ciclo con un nodo adicional conectado
edges_A = [(0, 1), (1, 2), (2, 0), (0, 3), (3, 4)]  # Un triángulo con un nodo adicional
G_A.add_edges_from(edges_A)

# Grafo B: Cambia la estructura pero mantiene las conexiones
edges_B = [(0, 1), (1, 2), (1, 3), (3, 4), (2, 4)]  # Un nodo central conectado a varios
G_B.add_edges_from(edges_B)

# Grafo C: Copia de A
edges_C = edges_A  # C es idéntico a A
G_C.add_edges_from(edges_C)

# Dibujar los tres grafos
plt.figure(figsize=(10, 5))

plt.subplot(131)
nx.draw_networkx(G_A, with_labels=True, node_color='lightblue')
plt.title("Grafo A")

plt.subplot(132)
nx.draw_networkx(G_B, with_labels=True, node_color='lightgreen')
plt.title("Grafo B")

plt.subplot(133)
nx.draw_networkx(G_C, with_labels=True, node_color='lightcoral')
plt.title("Grafo C")

plt.tight_layout()
plt.show()
````

````python
# Comprobar si A y B son isomorfos según faster_could_be_isomorphic
print("A y B son isomorfos según faster_could_be_isomorphic:", faster_could_be_isomorphic(G_A, G_B))

# Comprobar si A y B son isomorfos según is_isomorphic
print("A y B son isomorfos según is_isomorphic:", is_isomorphic(G_A, G_B))

# Comprobar si A y C son isomorfos según is_isomorphic
print("A y C son isomorfos según is_isomorphic:", is_isomorphic(G_A, G_C))
````

### Pregunta 4: PageRank


````python
# Recrea el grafo aquí
````

````python
import networkx as nx
import matplotlib.pyplot as plt

# Crear un grafo dirigido
G = nx.DiGraph()

# Añadir las aristas
edges = [(0, 2), (0, 4), (0, 5),
         (1, 4), (1, 5),
         (2, 3), (2 ,4),
         (3, 6),
         (4, 5),
         (6, 3),
         (8, 2), (8, 6), (8, 7)]

G.add_edges_from(edges)

# Definir las posiciones de los nodos manualmente para que se asemeje a la imagen del ejericio
pos = {0: (2, 2),
       1: (0, 2),
       2: (2, 1),
       3: (3, 2),
       4: (0.8, 1.2),
       5: (0.8, 2.7),
       6: (2.2, 2.7),
       7: (0, 3),
       8: (1, 2)}

# Dibujar el grafo con las posiciones especificadas
nx.draw(G, pos, with_labels=True, node_size=700, node_color="lightblue", arrows=True)

plt.show()
````

````python
# Calcula pagerank aquí
````

````python
# Calcular el PageRank
pagerank_values = nx.pagerank(G)

# Mostrar los valores de PageRank por nodo
print("Valores de PageRank por nodo:")
for node, value in pagerank_values.items():
    print(f"Nodo {node}: {value}")
````

```
PageRank mide la importancia de los nodos en un grafo, basado en la
cantidad y relevancia de los enlaces que reciben de otros nodos. La
idea es que un nodo es más importante si muchos otros nodos importantes
apuntan a él.

El proceso comienza asignando un valor inicial de PageRank igual a
todos los nodos. Luego, este valor se distribuye entre los nodos a
los que apunta cada uno. Este intercambio de valores ocurre varias
veces, hasta que los valores se estabilizan. Además, el algoritmo
incluye un "factor de amortiguación" que simula que un usuario puede
saltar a cualquier nodo al azar, lo que asegura que todos los nodos
reciban al menos algo de PageRank.

En mi grafo, los valores de PageRank obtenidos son:

Nodo 0: 0.0316
Nodo 2: 0.0496
Nodo 4: 0.0751
Nodo 5: 0.1179
Nodo 1: 0.0316
Nodo 3: 0.3142
Nodo 6: 0.3077
Nodo 8: 0.0316
Nodo 7: 0.0406

El nodo 3 tiene el valor más alto (0.3142), lo que indica que es el más
influyente dentro del grafo. Esto ocurre porque recibe enlaces de otros
nodos importantes como el 6, que también tiene un valor alto (0.3077).
En cambio, nodos como el 0, 1 y 8 tienen valores muy bajos, porque están
más desconectados o no reciben tantos enlaces de nodos relevantes.

El valor de PageRank de un nodo no solo depende de cuántos enlaces entrantes
tenga, sino de la importancia de los nodos que apuntan a él. En este caso,
el nodo 3 se ve beneficiado por estar conectado con nodos que a su vez son
importantes, como el nodo 6.
```

````python
import networkx as nx
import matplotlib.pyplot as plt

# 1. Cargar el dataset
# El archivo "facebook_combined.txt" contiene las relaciones entre usuarios de Facebook en forma de aristas
G = nx.read_edgelist("facebook_combined.txt", create_using=nx.Graph(), nodetype=int)

# 2. Calcular el PageRank
pagerank_values = nx.pagerank(G)

# 3. Obtener los tamaños de los nodos proporcionalmente a los valores de PageRank
node_sizes = [v * 10000 for v in pagerank_values.values()]  # Multiplicamos para ajustar tamaño

# 4. Mostrar el grafo
plt.figure(figsize=(14, 12))

# Posición de los nodos utilizando un layout spring
pos = nx.spring_layout(G)

# Dibujar el grafo con tamaños de nodos según PageRank
nx.draw_networkx(
    G, pos,
    node_size=node_sizes,  # Tamaño de los nodos
    node_color=list(pagerank_values.values()),  # Colores basados en PageRank
    cmap=plt.cm.viridis,  # Escala de colores
    with_labels=False,
    edge_color="gray"
)

plt.title("Grafo de Facebook con tamaños de nodos basados en PageRank")
plt.colorbar(plt.cm.ScalarMappable(cmap=plt.cm.viridis), label="PageRank")
plt.show()
````
