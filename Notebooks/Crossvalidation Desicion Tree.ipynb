```python
Dataset = Dataset.replace(to_replace ="Iris-setosa",          value ="0")
Dataset = Dataset.replace(to_replace ="Iris-versicolor",      value ="1")
Dataset = Dataset.replace(to_replace ="Iris-virginica",       value ="2")
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state = 100)
```

```python
clf_gini_PROFUNDIDAD_5_02.fit(X_train, Y_train)
```

```python
y_pred_ginI_5 = clf_gini_PROFUNDIDAD_5_02.predict(X_test)
```

```python
print ("Accuracy : ", accuracy_score(Y_test,y_pred_ginI_5)*100)         # Evaulating predictions with test labels
print ("Report : ",  classification_report(Y_test, y_pred_ginI_5))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.datasets import load_iris
from sklearn.metrics import accuracy_score, mean_squared_error, r2_score

# Cargar datos de ejemplo
data = load_iris()
X = data.data[:, :2]  # Usar solo las primeras dos características para simplificar la visualización
Y = data.target

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)

# Definir los hiperparámetros a ajustar utilizando Grid Search Cross-Validation
param_grid = {'max_depth': [5, 10, 15]}

# Crear el modelo DecisionTreeClassifier
model = DecisionTreeClassifier(random_state=0)

# Usar validación cruzada con GridSearchCV donde cv=3
grid_search = GridSearchCV(model, param_grid, cv=3, scoring='neg_mean_squared_error')
grid_search.fit(X_train, Y_train)

# Obtener los mejores hiperparámetros encontrados
best_params = grid_search.best_params_

# Entrenar el modelo con los mejores hiperparámetros
best_model = DecisionTreeClassifier(**best_params, random_state=0)
best_model.fit(X_train, Y_train)

# Realizar predicciones en el conjunto de prueba
predictions = best_model.predict(X_test)

# Calcular métricas de evaluación
mse = mean_squared_error(Y_test, predictions)
r2 = r2_score(Y_test, predictions)
accuracy = accuracy_score(Y_test, predictions)

print(f'MSE: {mse}')
print(f'R²: {r2}')
print(f'Accuracy: {accuracy}')
print(f'Best Parameters: {best_params}')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.datasets import load_iris
from sklearn.metrics import accuracy_score, mean_squared_error, r2_score, confusion_matrix
import seaborn as sns

# Cargar datos de ejemplo
data = load_iris()
X = data.data  # Usar todas las características disponibles
Y = data.target

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=0)

# Definir los hiperparámetros a ajustar utilizando Grid Search Cross-Validation
hyperparameters = [{'max_depth': [1, 2, 3, 4, 5], 'criterion': ['gini', 'entropy']}]

# Crear el modelo DecisionTreeClassifier
model = DecisionTreeClassifier(random_state=0)

# Usar validación cruzada con GridSearchCV donde cv=3
grid_search = GridSearchCV(model, hyperparameters, cv=3, scoring='neg_mean_squared_error')
grid_search.fit(X_train, Y_train)

# Obtener los mejores hiperparámetros encontrados
best_params = grid_search.best_params_

# Entrenar el modelo con los mejores hiperparámetros
best_model = DecisionTreeClassifier(**best_params, random_state=0)
best_model.fit(X_train, Y_train)

# Obtener el valor de Gini o Entropy del nodo raíz
root_impurity = best_model.tree_.impurity[0]
criterion_used = best_model.criterion

print(f'Criterio utilizado: {criterion_used}')
print(f'Valor de Gini o Entropy en el nodo raíz: {root_impurity}')

# Realizar predicciones en el conjunto de prueba
predictions = best_model.predict(X_test)

# Calcular métricas de evaluación
mse = mean_squared_error(Y_test, predictions)
r2 = r2_score(Y_test, predictions)
accuracy = accuracy_score(Y_test, predictions)

print(f'MSE: {mse}')
print(f'R²: {r2}')
print(f'Accuracy: {accuracy}')
print(f'Best Parameters: {best_params}')

# Calcular y mostrar la matriz de confusión
conf_matrix = confusion_matrix(Y_test, predictions)
sns.heatmap(conf_matrix, annot=True, cmap='Blues', fmt='g')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
